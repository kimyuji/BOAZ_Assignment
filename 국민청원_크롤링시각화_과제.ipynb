```python
import pandas as pd
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver

#headless: driver.get()을 해도 창이 뜨지 않게 하는 옵션
options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')

driver = webdriver.Chrome(executable_path='C:/Users/USER/Desktop/BOAZ/BOAZ 시각화 세션/chromedriver', chrome_options= options)

import time
import re
import datetime
from tqdm import tqdm
```

```python
#list 초기화
num=[]
cat=[]
title=[]
date=[]
pop=[]

#url 돌려가며 정보 가져오기
for i in tqdm(range(1,2001)):
    url='https://www1.president.go.kr/petitions/?c=0&only=2&page={}&order=2'.format(i)
    time.sleep(1)
    driver.get(url)
    time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    #번호
    n=soup.select('.bl_no')[7:]
    for i in range(0,len(n)):
        num.append(re.search('</span>(.+)</div>',str(n[i])).group(1)) 
    #분류
    c=soup.select('.bl_category.ccategory.cs.wv_category')[5:]
    for i in range(0,len(c)):
        cat.append(re.search('</span>(.+)</div>',str(c[i])).group(1))
    #제목
    t=soup.select('a.cb.relpy_w')[14:]
    for i in range(0,len(t)):
        title.append(re.search('</span>(.+)</a>',str(t[i])).group(1))
    #청원만료일
    d=soup.select('.bl_date.light')[14:]
    for i in range(0,len(d)):
        date.append(re.search('</span>(.+)</div>',str(d[i])).group(1))
    #참여인원
    p=soup.select(".bl_agree.cs")[5:]
    for i in range(0,len(p)):
        pop.append(re.search('</span>(.+)</div>',str(p[i])).group(1))
```

##### 청원 페이지에서 조금 이따가 시도하라는 에러 매세지가 반복적으로 나타났습니다
##### 추가적인 코드로 2000페이지까지 다 돌렸읍니다...^^ 

```python
data={'번호': num,
     '분류': cat,
     '제목': title,
     '청원 만료일': date,
     '참여인원': pop}
df = pd.DataFrame(data)
df.to_csv('Result1.csv') 
df
```

### 데이터 전처리

```python
df.info()
```

```python
#청원 만료일 전처리
for i in range(0,df.shape[0]):
    df.loc[i,'청원 만료일']=datetime.datetime.strptime(df.loc[i,'청원 만료일'], '%Y-%m-%d')
```

```python
#참여인원 전처리
for i in range(0,df.shape[0]):
    text=re.search('(.+)명',str(df.loc[i,'참여인원'])).group(1)
    text=re.sub('[,]', '', text) #쉼표 제거
    df.loc[i,'참여인원']=int(text)
```

```python
#CSV파일로 내보내기
df.to_csv('Result2.csv') 
```
