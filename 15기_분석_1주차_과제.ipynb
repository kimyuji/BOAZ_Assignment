## 변수에 따른 평균 집값 비교

```python
df.groupby('bedrooms')['price'].agg([('침실','mean')]).plot(kind='bar', figsize=(12, 6), color='lightpink')
plt.title('침실 수에 따른 평균 집값', fontsize='15')
plt.xlabel('침실 개수')
plt.show()
```

```python
df.groupby('bathrooms')['price'].agg([('욕실','mean')]).plot(kind='bar', figsize=(16, 6), color='lightblue')
plt.title('욕실 수에 따른 평균 집값', fontsize='15')
plt.xlabel('욕실 개수')
plt.show()
```

```python
df.groupby('floors')['price'].agg([('층 수','mean')]).plot(kind='bar', figsize=(12, 6), color='darkseagreen')
plt.title('층 수에 따른 평균 집값', fontsize='15')
plt.xlabel('층 수')
plt.show()
```

```python
df.groupby('waterfront')['price'].agg([('물가','mean')]).plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('물가(waterfront)에 따른 평균 집값', fontsize='15')
plt.xlabel('물가 여부')
plt.show()
```

```python
df.groupby('view')['price'].agg([('전망','mean')]).plot(kind='bar', figsize=(12, 6), color='lightblue')
plt.title('전망에 따른 평균 집값', fontsize='15')
plt.xlabel('전망')
plt.show()
```

```python
df.groupby('condition')['price'].agg([('상태','mean')]).plot(kind='bar', figsize=(12, 6), color='salmon')
plt.title('상태에 따른 평균 집값', fontsize='15')
plt.xlabel('상태 점수')
plt.show()
```

```python
df.groupby('grade')['price'].agg([('점수','mean')]).plot(kind='bar', figsize=(16, 6), color='violet')
plt.title('점수에 따른 평균 집값', fontsize='15')
plt.xlabel('점수')
plt.show()
```

- 침실 수, 욕실 수, 층 수가 많을 수록 집값이 높은 경향을 보인다.
- 집근처에 물가가 있는 경우 평균 집값이 높다. 
- 전망이 좋을 수록, 상태가 수록, 점수가 높을수록 집값이 높은 경향을 보인다. 
- 하지만, 침실 수와 욕실 수, 층 수는 너무 많게 되면 오히려 평균 집값이 낮아진다. 


```python
df['color']=np.nan
for i in range(len(df['price'])):
    if list(df['price'])[i] >=645000: df.loc[i,'color']='red' #집값 상위 25%
    elif list(df['price'])[i] >=450000: df.loc[i,'color']='orange' #집값 상위 25%~50%
    elif list(df['price'])[i] >=322000: df.loc[i,'color']='yellow' #집값 상위 50%~75%
    else: df.loc[i,'color']='lightgrey' # 집값 하위 25%
```

### sqft_living과 sqft_lot 변수 알아내기
EDA를 하면서 sqft_living과 sqft_lot이 정확이 어떤 것을 의미하는지 궁금했습니다. 
검색해본 결과:
- living : 생활하는 공간, heated or cooled가 가능한 공간이며, 모든 층을 포함한 수치이다. (다락방, 창고, 테라스 등은 제외(
- lot : 땅(부지) 사이즈이며, 마당을 포함한 것이다. 

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
plt.title('층수에 따른 sqft_living & sqft_lot', fontsize='15')
plt.figure(figsize=(25, 10))
ax0.plot(df.groupby('floors')['sqft_living'].agg([('층 수','mean')]), 'r-', label="y0")
ax0.set_ylabel("sqft_living")
ax0.set_ylim([0, 17000])
ax0.grid(False)
ax1.plot(df.groupby('floors')['sqft_lot'].agg([('층 수','mean')]), 'b-', label="y1")
ax1.set_ylabel("sqft_lot")
ax1.grid(False)
ax1.set_ylim([0, 17000])
ax0.set_xlabel("층 수")
plt.show()
```

sqft_living은 층수를 포함한 면적이기 때문에 층 수가 많을 수록 sqft_living 값이 증가하고,
sqft_lot은 층수를 포함하지 않기 때문에 층수와는 값이 상관이 없어 보인다. 

```python
plt.figure(figsize=(12, 10))         
df.groupby('floors')[['bathrooms','bedrooms']].agg([('침실 수','median')]).plot(kind='bar', figsize=(12, 6), color=['lightblue','lightpink'], stacked=True)
plt.xlabel("층 수")                  
plt.ylabel("욕실 수+침실 수")                   
plt.show()    
```

층수가 많을수록 침실 수와 욕실 수가 많지 않을까?! 하였는데 그것도 아닌거 같습니다...ㅠㅠ <br/>
층수와 욕실 수는 양의 관계가 있어보입니다..

```python
plt.figure(figsize=(12, 10))         
plt.scatter(df.grade, df.condition) 
plt.xlabel("점수")                  
plt.ylabel("상태")                   
plt.show()                      
```

```python
#grade와 condition의 상관관계
import scipy.stats as stats
stats.pearsonr(df.grade, df.condition)
```

상태가 좋을 수록 점수도 좋지 않을까? 라고 생각했으나 둘은 상관관계가 거의 없다는 것을 알게 되었습니다...

```python
import datetime
for i in range(0,df.shape[0]):
    df.loc[i,'date']=datetime.datetime.strptime(df.loc[i,'date'], '%Y%m%dT000000')
```

```python
df.groupby('date')['price'].agg([('가격','mean')]).plot(kind='line', figsize=(16, 6), color='violet')
plt.title('날짜에 따른 평균 집값', fontsize='15')
plt.xlabel('년도')
plt.show()
```

시간에 따라 집값이 크게 오르거나 하진 않은 것 같습니다. 다만 중간에 치솟아 있는 이상치의 경우 그 날짜에 고가의 집이 하나 거래됐기 때문에 평균값이 크게 나타난 것으로 보입니다.
